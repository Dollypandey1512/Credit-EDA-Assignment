#### Converting DAYS_BIRTH column into AGE for readability

```python
# divide 'DAYS_BIRTH' by 365 for taking age
application_data['AGE']=round(application_data['DAYS_BIRTH']/365,2)
```

```python
# drop DAYS_BIRTH column
application_data=application_data.drop(['DAYS_BIRTH'],axis=1)
```

```python
application_data.head()
```

As AGE column have negative values we have to convert them in positive values by using .abs

```python
#converting negative value into positive
application_data['AGE']=application_data['AGE'].abs()
```

```python
application_data.head()
```

# Handling Outliers

```python
#finding outliers in all numerical columns
outliers=['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_GOODS_PRICE','AGE']
for i in outliers:
  q1=application_data[i].quantile(0.25)
  q3=application_data[i].quantile(0.75)
  iqr=q3-q1
  upper_limit=q3+(1.5*iqr)
  lower_limit=q1-(1.5*iqr)
  application_data=application_data.loc[(application_data[i]<upper_limit)&(application_data[i]>lower_limit)]

application_data.shape
```

```python
#percentage of rows lost
round((100-100*(len(application_data.index)/307511)),2)
```

We have lost around 27percent rows

```python
# Binning of AGE,AMT_INCOME_TOTAL,AMT_CREDIT,EXT_SOURCE column for further outliers
#creating three age group young mid age and senior citizen
```

```python
application_data['AGE'].describe()
```

```python
#define function for categorizing age group
def age_group(x):
  if (x>40):
     return'YOUNG'
  elif (x>=40 and x < 60):
     return 'MID AGE'
  else:
      return'SENIOR'
```

```python
# creating new column AGE_GROUP
application_data['AGE_GROUP']=application_data['AGE'].apply(age_group)
application_data.head()
```

```python
#creating credit amount group
round(application_data['AMT_CREDIT'].describe(),2)
```

```python
#define function for categorizing CREDIT_GROUP (High,Medium,Low)
def credit_group(x):
  if (x<500000):
    return 'Low'
  elif(x>=500000 and x<750000):
    return 'Medium'
  else:
      return 'High'
```

```python
#creating new column CREDIT_GROUP
application_data['CREDIT_GROUP']=application_data['AMT_CREDIT'].apply(credit_group)
application_data.head()
```

```python
#creating INCOME_GROUP (High,Medium,Low)
application_data['AMT_INCOME_TOTAL'].describe()
```

```python
#define function for categorizing INCOME_GROUP (High,Medium,Low)
def income_group(x):
  if (x<100000):
    return 'Low'
  elif (x>=100000 and x<200000):
    return 'Medium'
  else:
    return 'High'
```

```python
#creating column INCOME_GROUP with AMT_INCOME_TOTAL(Low,Medium,High)
application_data['INCOME_GROUP']=application_data['AMT_INCOME_TOTAL'].apply(income_group)
application_data.head()
```
